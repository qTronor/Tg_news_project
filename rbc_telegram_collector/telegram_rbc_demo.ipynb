```python
# 3) Создание строковой сессии (РЕШЕНИЕ ПРОБЛЕМЫ ЗАВИСАНИЯ)
# ⚠️ ВАЖНО: Если сбор зависает, это обычно из-за отсутствия авторизации
# Запустите эту ячейку для создания строковой сессии (потребуется телефон и код из Telegram)

from telethon import TelegramClient
from telethon.sessions import StringSession
import os

async def create_string_session():
    api_id = int(os.getenv("TG_API_ID"))
    api_hash = os.getenv("TG_API_HASH")
    
    print("Создание строковой сессии...")
    print("Вам будет запрошен номер телефона и код из Telegram")
    
    async with TelegramClient(StringSession(), api_id, api_hash) as client:
        session_string = client.session.save()
        print(f"\n{'='*60}")
        print(f"TG_STRING_SESSION={session_string}")
        print(f"{'='*60}\n")
        print("ВАЖНО: Сохраните это значение!")
        print("Добавьте в ячейку 2:")
        print(f'os.environ["TG_STRING_SESSION"] = "{session_string}"')
        return session_string

# Раскомментируйте следующую строку для создания сессии:
# await create_string_session()
```

```python
# 4) Запуск сбора
from collector.runner import load_config, collect_once
import asyncio
from pathlib import Path
from datetime import date
import logging

# Настройка логирования для отладки
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s %(levelname)s %(name)s: %(message)s"
)

cfg = load_config(Path("config.example.yaml"))
cfg.channels[0].since = date(2026, 1, 1)
cfg.channels[0].limit = 1

print("Начинаем сбор...")
print(f"Канал: {cfg.channels[0].name}")
print(f"Лимит: {cfg.channels[0].limit}")

# В Jupyter ноутбуках лучше использовать asyncio.run() для надежности
# или убедиться, что event loop уже запущен
try:
    # Проверяем, есть ли уже запущенный event loop
    loop = asyncio.get_event_loop()
    if loop.is_running():
        # Если loop уже запущен (в Jupyter), используем await напрямую
        results = await collect_once(cfg)
    else:
        # Если loop не запущен, используем asyncio.run()
        results = asyncio.run(collect_once(cfg))
except RuntimeError:
    # Если нет event loop, создаем новый
    results = asyncio.run(collect_once(cfg))

print(f"\nРезультат: {results}")
results
```
